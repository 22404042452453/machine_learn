```python
import pandas as pd
import numpy as np
df = pd.read_csv('BlackFriday.csv',sep =',')
my_group = df.groupby(by = ['Age',"Gender"], as_index = False).agg({"User_ID":'count'})
my_women = my_group.query('Gender == "F" & Age >"36"').sum()
my_man = my_group.query('Gender == "M" & Age >"26" & Age < "35"').sum()
round((my_man["User_ID"] + my_women["User_ID"])/my_group['User_ID'].sum(),4)


```
