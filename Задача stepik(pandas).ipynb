Вася составил раскладку по продуктам на один день (она на листе "Раскладка") с указанием названия продукта и его количества в граммах. Посчитайте 4 числа: суммарную калорийность и граммы белков, жиров и углеводов. Числа округлите до целых вниз и введите через пробел.


```python
import pandas as pd 

df = pd.read_excel("https://stepik.org/media/attachments/lesson/245290/trekking2.xlsx",sheet_name="Раскладка")
df1 = pd.read_excel("https://stepik.org/media/attachments/lesson/245290/trekking2.xlsx")

df.head()
df1.head()
```

```python
df1 = df1.rename(columns={"Unnamed: 0":"Продукт"})
df1.head()
```

```python
df_full = df1.merge(df)
df_full.fillna(0)
```

```python
df_full["calory"] = df_full["ККал на 100"] * df_full["Вес в граммах"]/100
df_full["belki"] = df_full["Б на 100"] * df_full["Вес в граммах"]/100
df_full["fats"] = df_full["Ж на 100"] * df_full["Вес в граммах"]/100
df_full["carbonion"] = df_full["У на 100"] * df_full["Вес в граммах"]/100



df_full
```
