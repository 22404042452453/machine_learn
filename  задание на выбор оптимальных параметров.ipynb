```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv("https://stepik.org/media/attachments/course/4852/training_mush.csv")

df.head()
```

```python
X = df.drop("class",axis =1)
X
```

```python
y = df['class']
y
```

```python
parametrs = {
'n_estimators':range(10,51,10),
'max_depth':range(1,13,2),
'min_samples_leaf':range(1,8),
'min_samples_split':range(2,10,2)   
}
```

```python
clf_rf = RandomForestClassifier(random_state=0)
grid_search_cv  = GridSearchCV(clf_rf,parametrs,n_jobs=-1,cv=3)
grid_search_cv.fit(X,y)
```

```python
grid_search_cv.best_params_
```

```python
best_rf = grid_search_cv.best_estimator_
best_rf.fit(X,y)
```

```python

df_feach = pd.DataFrame({"feature":list(X),
                  'feature_importances':best_rf.feature_importances_})

df_feach = df_feach.sort_values(by = 'feature_importances', ascending=False)
df_feach
```

```python
import seaborn as sns
np.random.seed(42)

sns.barplot(y = 'feature', x = 'feature_importances', data= df_feach,color='b', label = 'importances')

```

```python
df_test = pd.read_csv("https://stepik.org/media/attachments/course/4852/testing_mush.csv")

df_test
```

```python
best_rf.predict(df_test).sum()
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y,best_rf.predict(X))
```
