```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
%matplotlib inline 

df = pd.read_csv('https://stepik.org/media/attachments/course/4852/train_iris.csv',index_col=0)
df.head(10)
```

```python
X = df.drop(['species'], axis = 1)
y = df.species



```

```python
df1 = pd.read_csv('https://stepik.org/media/attachments/course/4852/test_iris.csv', index_col=0)
X_test = df1.drop(['species'], axis = 1)
y_test = df1.species

df1.head(20)
    
    
    
```

```python

max_depth = [i for i in range(1,101)]
my_train = []
my_test = []
for i in max_depth:
    clf2 = tree.DecisionTreeClassifier(criterion='entropy',max_depth=i)
    clf2.fit(X,y)
    my_train.append(clf2.score(X,y))
    my_test.append(clf2.score(X_test,y_test))
    
    
my_result = pd.DataFrame({'max_depth':max_depth,
                          'my_train':my_train,
                          'my_test':my_test})
    
my_result
    
```

```python
plt.figure(figsize = (13,10))

my_result.plot(x = 'max_depth', y =['my_train','my_test'])

plt.grid()
```
